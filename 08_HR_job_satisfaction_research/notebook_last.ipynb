C–æ–∑–¥–∞—ë–º —Å–ø–∏—Å–∫–∏ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
ohe_columns = ['dept',
               'last_year_promo',
               'last_year_violations']
ord_columns = ['level',
                'workload']
num_columns = ['salary',
               'employment_years',
               'supervisor_evaluation']
```

–°–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ohe_columns

```python
ohe_pipe = Pipeline([('simpleImputer_before_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                     ('ohe', OneHotEncoder(drop='first',
                                           sparse=False, 
                                           handle_unknown='ignore'
                                          ))])
```

–°–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ord_columns

```python
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[['junior', 'middle', 'sinior'],
                            ['low', 'medium', 'high']], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

C–æ–∑–¥–∞—ë–º –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

C–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
])
 
```

```python
param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LinearRegression() 
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },
 
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
#        'models__max_depth': range(2, 3),
#        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ RandomForestRegressor()
     {
        'models': [RandomForestRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(6, 10),
        'models__max_features': range(6, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
```

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ —Ç–∞–∫ –∂–µ, –∫–∞–∫ –∏ –≤ –ø–µ—Ä–≤–æ–π –∑–∞–¥–∞—á–µ: –≤—ã–ø–æ–ª–Ω–∏–º –ø–æ–¥–≥–æ—Ç–æ–≤–∫—É –≤ –ø–∞–π–ø–ª–∞–π–Ω–µ, –¥–æ–ø–æ–ª–Ω–∏–≤ –ø–∞–π–ø–ª–∞–π–Ω –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏. –ü—Ä–∏ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–∏ —É—á—Ç–µ–º –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –º–æ–¥–µ–ª–µ–π –∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–≤–∞ –∫–æ–¥–∏—Ä–æ–≤—â–∏–∫–∞.

```python
ohe_columns = ['dept',
               'last_year_promo',
               'last_year_violations']
ord_columns = ['level',
                'workload']
num_columns = ['salary',
               'employment_years',
               'supervisor_evaluation',
               'job_satisfaction_rate'
               ]
```

```python
label_encoder = LabelEncoder()
y_train_new = label_encoder.fit_transform(y_train_new)
y_test_new = label_encoder.transform(y_test_new)
```

–°–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ohe_columns

```python
ohe_pipe = Pipeline([('simpleImputer_before_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                     ('ohe', OneHotEncoder(drop='first',
                                           sparse=False, 
                                           handle_unknown='ignore'
                                          ))])
```

–°–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ord_columns

```python
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[['junior', 'middle', 'sinior'],
                            ['low', 'medium', 'high']], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

C–æ–∑–¥–∞—ë–º –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

C–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LogisticRegression(random_state=RANDOM_STATE))
])
```

<b>–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏‚ö†Ô∏è:</b>
–í—Å–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ —Å—Ç–æ–∏—Ç –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –≤ –Ω–∞—á–∞–ª–µ –ø—Ä–æ–µ–∫—Ç–∞ - —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –ø–æ–ª—É—á–∞—Ç–µ–ª—å —Ç–≤–æ–µ–≥–æ –æ—Ç—á—ë—Ç–∞ —Å–º–æ–∂–µ—Ç —É–∑–Ω–∞—Ç—å –æ –ø—Ä–æ–±–ª–µ–º–∞—Ö —Å –µ–≥–æ –æ–∫—Ä—É–∂–µ–Ω–∏–µ–º —Å—Ä–∞–∑—É –ø–µ—Ä–µ–¥ —Ä–∞–±–æ—Ç–æ–π —Å –ø—Ä–æ–µ–∫—Ç–æ–º, –∞ –Ω–µ –ø–æ—Å—Ä–µ–¥–∏ –∏–ª–∏ –≤ –∫–æ–Ω—Ü–µ –ø—Ä–æ–µ–∫—Ç–∞.

<div class="alert alert-success">
    
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ v.2 <a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> –£—á—Ç–µ–Ω–æ.

```python
param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression() 
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE)],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },
 
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(5, 7),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    {
        'models': [RandomForestClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(5, 7),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
```

<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"></h2>

    
<b>–ù–∞ –¥–æ—Ä–∞–±–æ—Ç–∫—É‚ùå:</b> –£ `LogisticRegression` –Ω–µ –∑–∞–¥–∞–Ω `random_state`.

```python
grid_search = GridSearchCV(
    pipe_final,
    param_grid,
    n_jobs=-1,
    cv=5,
    scoring='roc_auc'
)
grid_search.fit(X_train_new, y_train_new)
```

```python
print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', grid_search.best_estimator_)
print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', grid_search.best_score_)
```

```python
ohe_columns = ['last_year_promo',
               'last_year_violations']
ord_columns = ['level',
                'workload']
num_columns = ['salary', 
               'supervisor_evaluation',
               'employment_years',
              'job_satisfaction_rate']
```

```python
X_train_new_sorted = train_quit.drop(['quit', 'dept'], axis=1)
X_test_new_sorted = test_features.drop(['quit', 'dept'], axis=1)
X_train_new_sorted.shape, X_test_new_sorted.shape, y_train_new.shape, y_test_new.shape
```

```python
ohe_pipe = Pipeline([('simpleImputer_before_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                     ('ohe', OneHotEncoder(drop='first',
                                           sparse=False, 
                                           handle_unknown='ignore'
                                          ))])
```

```python
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'sinior'],
                    ['low', 'medium', 'high']
                           ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
data_preprocessor_ = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

```python
pipe_final_ = Pipeline([
    ('preprocessor', data_preprocessor_),
    ('models', LogisticRegression(random_state=RANDOM_STATE))
])
```


ROC-AUC –æ—Ç–ª–∏—á–Ω–æ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Å –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–æ–π.¬∂

```python
grid_search = GridSearchCV(
    pipe_final_,
    param_grid,
    n_jobs=-1,
    cv=5,
    scoring='roc_auc'
)
grid_search.fit(X_train_new_sorted, y_train_new)
```

```python
param_grid = [
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE)],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(6, 8),
        'models__max_features': range(7, 9),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },    
    {
        'models': [RandomForestClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(6, 7),
        'models__max_features': range(4, 9),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
```
